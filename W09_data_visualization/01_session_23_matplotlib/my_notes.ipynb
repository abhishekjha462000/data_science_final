## `Line plot`

    Line plot is the simplest type of data plot. It is used for bivariate analysis. We can have numerical-numerical,
    numerical-categorical kind of bivariate analysis using line plot.
    
    It is also important that we have the data sorted in the horizontal direction.

```python
prices = [45000, 23000, 34000, 45000 , 23000]
years = [2003, 2005, 2007, 2009, 2011]
```

```python
plt.figure(figsize = (20, 20))


plt.plot(years, prices, color = 'green', marker = 'D', linewidth = 2, markersize = 20)


plt.grid()
```

## `plotting graph using pandas dataframe`

```python
df = pd.read_csv('sharma-kohli.csv')
df.rename(columns = {'RG Sharma' : 'rg_sharma', 'V Kohli' : 'v_kohli'}, inplace = True)
df.head()
```

```python
plt.figure(figsize = (20, 20))


plt.plot(df['index'], df['v_kohli'], color = 'purple', marker = '+', linewidth = 3, markersize = 30, linestyle = '-.',
        label = 'virat_kohli')


plt.xlabel('years')
plt.ylabel('runs')
plt.legend(loc = 'upper left')
plt.show()
```

```python
plt.figure(figsize = (20, 20))


plt.plot(df['index'], df['v_kohli'], color = 'purple', marker = '+', linewidth = 1.5, markersize = 30, linestyle = '-.',
        label = 'virat_kohli')
plt.plot(df['index'], df['rg_sharma'], color = 'magenta', marker = 'o', linewidth = 2.5, markersize = 10, linestyle = 'dashed',
        label = 'rg_sharma')


plt.xlabel('years')
plt.ylabel('runs')
plt.legend(loc = 'upper left')
plt.show()
```

### `Limiting the axes`

```python
plt.figure(figsize = (20, 20))


plt.plot(df['index'], df['v_kohli'], color = 'purple', marker = '+', linewidth = 1.5, markersize = 30, linestyle = '-.',
        label = 'virat_kohli')
plt.plot(df['index'], df['rg_sharma'], color = 'magenta', marker = 'o', linewidth = 2.5, markersize = 10, linestyle = 'dashed',
        label = 'rg_sharma')


plt.xlabel('years')
plt.ylabel('runs')
plt.xlim([2008, 2012])  # Here we limited the x-axis as we were only interested in this range
plt.legend(loc = 'upper left')
plt.grid()
plt.show()
```

### `We can also specify the size of each marker by using the size parameter. We can provide it with a series.`

```python
tips = sns.load_dataset('tips')
tips.head()
```

    The keyword argument s controls the size of each marker. We can provide it with a series the values of which will
    be used to give the size of the marker.
    
    The higher the value the larger the marker. The lower the value the smaller the marker.
    
    In a way, it can help us make 3D plot(how?). Because now we are able to convey 3 pieces of information together.

```python
plt.figure(figsize = (20, 20))


plt.scatter(tips['total_bill'], tips['tip'], s = tips['size'] * 100, marker = '$TIP$', label = 'total_bill vs tip')
# s = tips['size'] * 100 specifies the size of the marker. The greater the value the greater the size of the marker.


plt.xlabel('total_bill')
plt.ylabel('tip')
plt.grid()
plt.legend()
plt.show()
```

    The above graph is a 3D graph as we are conveying 3 pieces of information together. The position of the point refers to
    the total bill and the amount given as tip. However, we also have a 3rd piece of information which is the size of the 
    point. The larger the marker at a particular co-ordinate the higher its size.
    
    
    
    Also, from the above graph we can see that their is a positive correlation between the total bill and the tip amount.
    The lower the bill the less the tip and the higher the bill the higher the tip.

# `Scatter plot using plt.plot()`

    We can also plot a scatter plot using the plt.plot() method. We can achive this in 3 ways.
    
    The first way is that we provide the line-format with no linestyle.
    
    
    But the main question is why should we use plt.plot() when pyplot api gives us scatter() method?
    
    The simple answer to this question is that plt.plot() is much faster than plt.scatter() method.
    
    However, we cannot do the c-parameter customization in plt.plot() as we did in plt.scatter() method.

```python
prices = [45000, 23000, 34000, 45000 , 23000]
years = [2003, 2005, 2007, 2009, 2011]
```

```python
plt.figure(figsize = (20, 20))


plt.plot(years, prices, linestyle = '', color = 'green', marker = 'D', linewidth = 2, markersize = 20)


plt.grid()
```

    The second way is that while providing format string we do not specify the linestyle.

```python
plt.figure(figsize = (10, 10))


plt.plot(years, prices, 'D', color = 'green', markersize = 15)


plt.grid()
```

    The third way is that we provide the linewidth very very small(just like epsilon).

```python
plt.figure(figsize = (20, 20))


plt.plot(years, prices, linestyle = '-.', linewidth = 0.00005, marker = 'D', color = 'green', markersize = 30)


plt.grid()
```
