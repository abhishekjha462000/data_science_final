### **`Problem-1:`** Draw a line plot of which, the x-axis is the "Year" and the y-axis is sum of "PM2.5" of two countries Iran and China.

```python
df = pd.read_csv('list_of_most_polluted_cities_by_particulate_matter_concentration.csv')
df.head()
```

```python
df.drop(columns = ['Unnamed: 0'], inplace = True)

df
```

```python
mask = df['Country'] == 'Iran'
iran_series = df[mask]
iran_series
```

```python
mask = df['Country'] == 'China'
china_series = df[mask]
china_series
```

```python
iran = iran_series.groupby('Year')['PM2.5'].sum()
iran
```

```python
china = china_series.groupby('Year')['PM2.5'].sum()
china
```

```python
plt.figure(figsize = (20, 20))

plt.plot(iran.index, iran.values, label = 'iran', linestyle = '-.', color = 'green', marker = 'D', 
         linewidth = 0.7, markersize = 10)
plt.plot(china.index, china.values, label = 'china', linestyle = '--', color = '#4567ef', marker = '+', 
         linewidth = 1, markersize = 20)

plt.xlabel('years')
plt.ylabel('sum_pm_2.5')
plt.legend(loc = 'upper left')
plt.grid()
plt.show()
```

### **`Problem-2:`** Draw a histogram of the  column "PM10" of which the y-axis represents the probability (see the documentation how to draw the probability).

```python
df
```

```python
plt.figure(figsize = (20, 20))

plt.hist(df['PM10'], density = True, edgecolor = 'gold', color = '#a9a998', bins = 50, label = 'PM10 density histogram')

plt.xlabel('PM10')
plt.ylabel('probability')
plt.title('PM10 probability density function')
plt.legend()
plt.grid()
plt.show()
```

### `Problem-6`
Show a line plot of Total Profit for each month with below styling.
* Dotted Line
* Line Color Blue
* Show Legend at top left

```python
df = pd.read_csv('company_sales_data.csv')
df
```

```python
plt.figure(figsize = (20, 20))

plt.plot(df['month_number'], df['total_profit'], color = 'blue', label = 'total_profit', marker = 'o',
         markerfacecolor = 'white', markeredgecolor = 'gold', markersize = 20)

plt.legend(loc = 'upper left')
plt.xlabel('month')
plt.ylabel('total_profit')
plt.xticks(np.arange(1, 13), 
           labels = ['Jan', 'Feb', 'March', 'April', 'May', 'June', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
           rotation = 'vertical'
          )
plt.show()
```

### `Problem-8` Multiline Plot of all products sales.
