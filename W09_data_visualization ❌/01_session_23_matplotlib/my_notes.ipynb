### `We can also specify the size of each marker by using the size parameter. We can provide it with a series.`

```python
tips = sns.load_dataset('tips')
tips.head()
```

    The keyword argument s controls the size of each marker. We can provide it with a series the values of which will
    be used to give the size of the marker.
    
    The higher the value the larger the marker. The lower the value the smaller the marker.
    
    In a way, it can help us make 3D plot(how?). Because now we are able to convey 3 pieces of information together.

```python
plt.figure(figsize = (20, 20))


plt.scatter(tips['total_bill'], tips['tip'], s = tips['size'] * 100, marker = '$TIP$', label = 'total_bill vs tip')
# s = tips['size'] * 100 specifies the size of the marker. The greater the value the greater the size of the marker.


plt.xlabel('total_bill')
plt.ylabel('tip')
plt.grid()
plt.legend()
plt.show()
```

    The above graph is a 3D graph as we are conveying 3 pieces of information together. The position of the point refers to
    the total bill and the amount given as tip. However, we also have a 3rd piece of information which is the size of the 
    point. The larger the marker at a particular co-ordinate the higher its size.
    
    
    
    Also, from the above graph we can see that their is a positive correlation between the total bill and the tip amount.
    The lower the bill the less the tip and the higher the bill the higher the tip.

# `Scatter plot using plt.plot()`

    We can also plot a scatter plot using the plt.plot() method. We can achive this in 3 ways.
    
    The first way is that we provide the line-format with no linestyle.
    
    
    But the main question is why should we use plt.plot() when pyplot api gives us scatter() method?
    
    The simple answer to this question is that plt.plot() is much faster than plt.scatter() method.
    
    However, we cannot do the c-parameter customization in plt.plot() as we did in plt.scatter() method.

```python
prices = [45000, 23000, 34000, 45000 , 23000]
years = [2003, 2005, 2007, 2009, 2011]
```

```python
plt.figure(figsize = (20, 20))


plt.plot(years, prices, linestyle = '', color = 'green', marker = 'D', linewidth = 2, markersize = 20)


plt.grid()
```

    The second way is that while providing format string we do not specify the linestyle.

```python
plt.figure(figsize = (10, 10))


plt.plot(years, prices, 'D', color = 'green', markersize = 15)


plt.grid()
```

    The third way is that we provide the linewidth very very small(just like epsilon).

```python
plt.figure(figsize = (20, 20))


plt.plot(years, prices, linestyle = '-.', linewidth = 0.05, marker = 'D', color = 'green', markersize = 30)


plt.grid()
```

# `bar plot`

    The bar plot can be used for univariate as well as bivaiate analysis. Generally it is used for numerical vs 
    categorical analysis.
    
    The categorical quantity is placed on the x axis and the numerical quantity is placed on the axis.
    
    We can have vertical as well as horizontal bar charts.
    
    We can also have simple bar charts, nested bar charts and stacked bar charts.

```python
colors = ['red', 'green', 'blue', 'pink']
children = [20, 34, 23, 12]
```

```python
plt.figure(figsize = (20, 20))
plt.bar(colors, children, color = colors, width = 0.5)
plt.show()
```

```python
df = pd.read_csv('batsman_season_record.csv')
df
```

```python
plt.figure(figsize = (20, 20))
plt.bar(df['batsman'], df['2016'])
# plt.grid()
plt.show()
```

But there is a problem in this approach. We cannot make stacked bar charts in matplotlib. Let us see what happens when we 
plot two bar charts together.

```python
plt.figure(figsize = (20, 20))
plt.bar(df['batsman'], df['2015'], label = '2015')
plt.bar(df['batsman'], df['2016'], label = '2016')
plt.legend()
# plt.grid()
plt.show()
```

We see that graphs overlapped each other. This reduces the readability. This is one of the biggest problems in matplotlib
pyplot api.

To resolve this, we need to set the width of each individual chart manually and see if it works. quite a tedious task.

## `setting xticks`

    Sometimes we want to change the label names on the x-axis. This is done using x_ticks() method.

```python
plt.figure(figsize = (20, 20))
plt.bar(df['batsman'], df['2016'], color = ['#455632', '#f3f', '#455668', '#234578', '#8923ef'])


plt.xticks(np.arange(df.index.shape[0]), 
           ['ab_de_villiers', 'da_warner', 'ms_dhoni', 'rg_sharma', 'v_kohli'], 
           rotation = 'vertical')


plt.show()
```

## `stacked bar chart`

    We can also stack bar charts on top of each other by using the bottom parameter.

```python
plt.figure(figsize = (20, 20))

plt.bar(df['batsman'], df['2015'])
plt.bar(df['batsman'], df['2016'], bottom = df['2015']) # The bottom of this chart will be at the top of the '2015'
plt.bar(df['batsman'], df['2017'], 
        bottom = df['2015'] + df['2016']) # The bottom of this chart will be at the top of '2015' + '2016'

plt.show()
```

# `histogram`

    The histogram is just the same as the bar chart.
    
    Bs fark ye he ki isme numerical data hi hoga x-axis pr and the whole dataset would be divided into bins.
    
    The height of each bar would be equal to the number of elements in a bin.

```python
np.random.seed(0)
data = np.random.randint(1, 100, (45,))
data
```

```python
plt.hist(data)
plt.show()
```

    All but the last (righthand-most) bin is half-open.  In other words, if *bins* is::
    [1, 2, 3, 4]
    then the first bin is ``[1, 2)`` (including 1, but excluding 2) and
    the second ``[2, 3)``.  The last bin, however, is ``[3, 4]``, which
    *includes* 4.

```python
plt.hist(data, bins = [10, 50, 80])
```

let us count how many elements are in [10, 50] and [50, 90].

```python
# number of elements in 10 and 50
count1 = sum([True if (10 <= ele and ele <50) else False for ele in data])
count1
```

```python
# number of elements in 10 and 50
count2 = sum([True if (50 <= ele and ele <=80) else False for ele in data])
count2
```

# `pie chart`

    pie chart is used for univariate analysis and bivariate analysis.
    
    It is used to find contribution on a standard scale.
    
    usually used for categorical vs numerical analysis.

```python
np.random.seed(0)
nums = pd.Series(np.random.randint(1, 6, (100,)))
nums
```

```python
freq = nums.value_counts()
freq
```

```python
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.2f}%  ({v:d})'.format(p=pct,v=val)
    return my_autopct

plt.pie(freq, labels = freq.index, autopct = make_autopct(freq))
plt.show()
```
