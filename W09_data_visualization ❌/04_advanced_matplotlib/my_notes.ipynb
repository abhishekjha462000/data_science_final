```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# `More on scatter plots`

    We can also make our scattered plots coloured. For that we can use the c-parameter, this parameter takes in 
    a color or a sequence of colors or a sequence of numbers.

```python
iris = sns.load_dataset('iris')
iris
```

```python
try:
    
#     plt.figure(figsize = (20, 20))

    plt.scatter(iris['sepal_length'], iris['sepal_width'], s = iris['petal_length'] * 35, c = iris['species'])

#     plt.show()
except Exception as ex:
    print(ex)
```

`The above code will not work as we have passed a categorical column to the c-parameter.`

    If we want that the c-paramter works, we first need to convert our categorical column into a numerical column and
    then pass it.

```python
species = iris['species'].unique()

specie_code_dict = {specie:index+1 for index, specie in enumerate(species)}

specie_code_dict
```

```python
plt.figure(figsize = (20, 20))

plt.scatter(iris['sepal_length'], 
            iris['sepal_width'], 
            s = iris['petal_length'] * 35, 
            c = iris['species'].apply(lambda x : specie_code_dict[x]), # we converted species into a numerical column in order
            # for it to be accepted by the c-parameter.
           )

plt.colorbar()
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.show()
```

    {'setosa': 1, 'versicolor': 2, 'virginica': 3}
    
    The color map and the above dictionary makes the graph understandable.

# `Annotation`

    We use plt.text() for annotation. This takes in a x-coordinate and y-coordinate and a text value to be placed at
    (x, y).

### `Example 1`

```python
x = [1, 2, 3, 4, 5]

y = [2, -3, 4, 12, 13]
```

```python
plt.scatter(x, y)

for i in range(len(x)):
    plt.text(x[i], y[i], (x[i], y[i]))

plt.grid()
plt.show()
```

### `Example 2`

```python
batter = pd.read_csv('batter.csv')
batter
```

```python
np.random.seed(0)
random_df = batter.head(100).sample(25)
random_df
```

```python
plt.figure(figsize = (20, 20))

plt.scatter(random_df['avg'], random_df['strike_rate'], s = random_df['runs'])

for i in range(random_df.shape[0]):
    plt.text(random_df['avg'].iloc[i], random_df['strike_rate'].iloc[i], random_df['batter'].iloc[i])

plt.xlabel('avg')
plt.ylabel('strike_rate')
plt.show()
```

# `lines`

    We can also plot lines on our plot. We have axvline() and axhline() methods which take in x-value and y-value
    respectively.

```python
x = [1, 2, 3, 4, 5]

y = [2, -3, 4, 12, 13]

plt.scatter(x, y)

plt.axvline(3.75, color = 'red')
plt.axhline(9, color = 'green')

for i in range(len(x)):
    plt.text(x[i], y[i], (x[i], y[i]))

plt.grid()
plt.show()
```

# `subplots()`

    We also have another methodology in matplotlib other than the pyplot api. This is the subplots() approach.
    
    The plt.subplots() methods takes in the number of rows and number of columns and provides us a set of axes
    in the form of ndarray along with a figure object.

## `Example 1`

```python
plt.subplots(nrows = 1, ncols = 1)
```

    As we can see above that we got a tuple.The first element of the tuple is the figure object and the second element
    of the tuple is the axes object.
    
    Here we got only 1 axes as the nrows and ncols were both equal to 1.

## `Example 2`

```python
fig, ax = plt.subplots(figsize = (20, 20), nrows = 2, ncols = 2)
```

    Here we got a figure object and a numpy array consisting of axes.

```python
type(fig)
```

```python
type(ax)
```

```python
ax.shape
```

    We can extract individual axes objects from the ax numpy array by using destructuring.

```python
[[ax1, ax2], [ax3, ax4]] = ax
```

```python
ax1, ax2, ax3, ax4
```

    We may now plot individual graphs on each of the axes.

## `Example 3`

```python
x = np.linspace(-10, 10, 100)
```

```python
fig, [ax1, ax2] = plt.subplots(figsize = (20, 20), nrows = 2, ncols = 1)

ax1.plot(x, x**2, label = 'square')
ax1.legend()
ax1.grid()
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('square function')

ax2.scatter(x, x ** 3, label = 'cube', color = 'green', marker = 'o')
ax2.legend()
ax2.grid()
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('cube function')

None
```

## `Example 4`

```python
batter
```

```python
fig, ax = plt.subplots(figsize = (20, 20), dpi = 100, nrows = 2, ncols = 2)

# axes 1 (lineplot of avg vs strike rate)
ax[0,0].plot(batter.sort_values(by = 'avg')['avg'], 
             batter.sort_values(by = 'avg')['strike_rate'], 
             label = 'avg_vs_strike_rate',
             color = 'green',
             marker = 'o',
             markerfacecolor = 'white',
             markeredgecolor = 'gold',
             markersize = 5,
             linewidth = 0.4
            )
ax[0, 0].legend()
ax[0, 0].grid()
ax[0, 0].set_xlabel('avg')
ax[0, 0].set_ylabel('strike_rate')
ax[0, 0].set_title('strike_rate vs avg')


# axes 2 (line plot of avg vs runs)
ax[0, 1].scatter(batter['avg'], batter['runs'], label = 'avg_vs_runs',
              color = 'blue',
              marker = '+',
             )

ax[0, 1].legend()
ax[0, 1].set_xlabel('avg')
ax[0, 1].set_ylabel('runs')
ax[0, 1].set_title('runs vs avg')


# axes 3 (pdf of runs)
ax[1, 0].hist(batter['runs'], edgecolor = 'red', color = '#8fafef', label = 'pdf_runs')
ax[1, 0].legend()
ax[1, 0].set_xlabel('runs')
ax[1, 0].set_ylabel('density')
ax[1, 0].set_title('pdf of runs')

# axes 4 (pdf of strike_rate)
ax[1, 1].hist(batter['strike_rate'], density = True, edgecolor = 'gold', color = '#afafaf', label = 'pdf_strike_rate')
ax[1, 1].legend()
ax[1, 1].set_xlabel('strike_rate')
ax[1, 1].set_ylabel('density')
ax[1, 1].set_title('probability density function of strike_rate')

None
```

# `add_subplot() approach`

    We can also make multiple graphs using the add_subplot() methods.

```python
x = np.linspace(-10, 10, 10000)
```

```python
fig = plt.figure(figsize = (20, 20), dpi = 100, facecolor = 'darkcyan')


# axes 1
fontdict = {'fontsize': 20, 'fontfamily':'serif'}
ax1 = fig.add_subplot(2, 1, 1, fc = 'lightblue')

ax1.plot(x, x ** 2, label = '$x^2$')

ax1.set_xlabel('x', fontdict)
ax1.set_ylabel('$x^2$', fontdict)
ax1.set_title('The graph of $x^2$', fontdict)
ax1.legend()





# axes 2

fontdict = {'fontsize': 20, 'fontfamily':'sans-serif', 'color' : 'white'}

ax2 = fig.add_subplot(2, 1, 2)

ax2.plot(x, x ** 3, label = '$x^3$')

ax2.set_xlabel('x', fontdict)
ax2.set_ylabel('$x^3$', fontdict)
ax2.set_title('The graph of $x^3$', fontdict)
ax2.grid()
ax2.legend()

None
```

# `3D plots`

    We can also plot 3D graphs using matplotlib. we have methods such as plot_surface(), scatter3D() and plot3D() to plot
    3D graphs in matplotlib.

```python
x = np.arange(1, 11)
x
```

```python
y = np.arange(2, 12)
```

```python
y
```

```python
X, Y = np.meshgrid(x, y)
```

```python
X
```

```python
Y
```

```python
Z = X ** 2 + Y ** 2
Z
```

```python
fig = plt.figure(figsize = (20, 20))

ax = fig.add_subplot(projection = '3d')

surface = ax.plot_surface(X, Y, Z, cmap = 'viridis')

ax.set_xlabel('x', fontdict = {'fontsize': 20})
ax.set_ylabel('y', fontdict = {'fontsize': 20})
ax.set_zlabel('z', fontdict = {'fontsize': 20})

ax.set_title('Graph of $X^2 + Y^2$', fontdict = {'fontsize': 20})

ax.view_init(elev = 20, azim = 45) # to set the view angle, we have elevation angle wrt to z-axis
fig.colorbar(surface)

None
```

## `Example 2`

```python
from math import pi

x = np.linspace(-2 * pi, 2 * pi, 100)
y = np.sin(x) ** 2 + x
```

```python
X, Y = np.meshgrid(x, y)
```

```python
Z = (np.exp(X) + np.exp(Y))
```

```python
fig = plt.figure(figsize = (20, 20))

ax = fig.add_subplot(projection = '3d')

surface = ax.plot_surface(X, Y, Z, cmap = 'GnBu')

ax.set_xlabel('x', fontdict = {'fontsize': 20})
ax.set_ylabel('y', fontdict = {'fontsize': 20})
ax.set_zlabel('z', fontdict = {'fontsize': 20})

ax.set_title('Graph of $Z = f(X, Y)$', fontdict = {'fontsize': 20})
ax.view_init(elev = 20, azim = 45)

fig.colorbar(surface)

None
```

# `contour plots`

    contour plots are a way of representing 3d plots on a 2d surface. The z values are represented by the contour lines.
    

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-20, 34, 100)
```

```python
X, Y = np.meshgrid(x, y)
```

```python
Z = X ** 2 + Y ** 2
```

```python
fig = plt.figure(figsize = (20, 20))

ax = fig.add_subplot()
contour_plot = ax.contour(X, Y, Z, cmap = 'viridis')

ax.set_xlabel('x', fontdict = {'fontsize': 20})
ax.set_ylabel('y', fontdict = {'fontsize': 20})
ax.set_title('Contour plot of $Z = f(X, Y)$', fontdict = {'fontsize': 20})

fig.colorbar(contour_plot)

None
```

### `filled contour plot`

    These are just the regular contour plot but filled with colors in between.

```python
fig = plt.figure(figsize = (10, 10))

ax = fig.add_subplot()
contour_plot = ax.contourf(X, Y, Z, cmap = 'viridis')

ax.set_xlabel('x', fontdict = {'fontsize': 20})
ax.set_ylabel('y', fontdict = {'fontsize': 20})
ax.set_title('Contour plot of $Z = f(X, Y)$', fontdict = {'fontsize': 20})

fig.colorbar(contour_plot)

None
```
