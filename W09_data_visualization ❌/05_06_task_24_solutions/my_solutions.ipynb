### **Problem-3:** Make a 3D *Surface* plot of this below mathematical equation.

$$z = |x| - |y|$$

````python
# code here
````

````python
x = np.linspace(-10, 10, 200)
y = np.linspace(-100, 100, 400)
````

````python
X,  Y = np.meshgrid(x, y)
````

````python
Z = np.abs(X) - np.abs(Y)
````

````python
fig = plt.figure(figsize = (20, 20))
fontdict = {'fontsize': '20'}

ax = fig.add_subplot(projection = '3d')

ax.plot_surface(X, Y, Z)

ax.set_xlabel('$X$', fontdict)
ax.set_ylabel('$Y$', fontdict)
ax.set_zlabel('$Z$', fontdict)

ax.set_title('graph of $z = |x| - |y|$', fontdict)
None
````

### **Problem-4:** Draw the 3D *Contour plot* of this below equation:

$$z = |x| - |y|$$

````python
# code here
````

````python
fig = plt.figure(figsize = (20, 20))
fontdict = {'fontsize': '20'}

ax = fig.add_subplot()

contour_plot = ax.contour(X, Y, Z, cmap = 'jet')

ax.set_xlabel('$X$', fontdict)
ax.set_ylabel('$Y$', fontdict)

fig.colorbar(contour_plot)

ax.set_title('comtour plot of $z = |x| - |y|$', fontdict)
None
````

### **Problem-5:** Draw a second type of Countour plot of the below equation:

````python
# code here
````

````python
fig = plt.figure(figsize = (20, 20))
fontdict = {'fontsize': '20'}

ax = fig.add_subplot()

contour_plot = ax.contourf(X, Y, Z, cmap = 'viridis')

ax.set_xlabel('$X$', fontdict)
ax.set_ylabel('$Y$', fontdict)

fig.colorbar(contour_plot)

ax.set_title('contour plot of $z = |x| - |y|$', fontdict)
None
````
