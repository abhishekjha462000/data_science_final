# `Dataframes`

    Dataframes are 2D data structures in pandas. A dataframe can be considered as a collection of one or more series.
    
    We can either create a dataframe using pd.Dataframe() method or by using the pd.read_csv() method. Although the
    latter is more common.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame([
    [100, 90, 9],
    [90, 80, 8],
    [110, 95, 15]
], columns = ['iq', 'marks', 'package'])

df
```

#### `we may also pass in a dictionary.`

```python
df = pd.DataFrame({
    'iq': [100, 90, 110],
    'marks': [90, 80, 95],
    'package': [9, 8, 15] 
})

df
```

#### isnull()

    This iterates over the whole dataframe and tells us whether a value is missing or not.
    By missing value, we mean NaN value. A value which is not NaN is called non-missing value.

```python
movies.isnull()
```

    We may then use the sum() method on the obtained boolean dataframe to get the total number of missing values
    per column.

```python
movies.isnull().sum()
```

    We can now use sum() method on the above obtained series object to get the total number of missing values in
    the entire dataframe.

```python
movies.isnull().sum().sum()
```

## `filtering dataframes`

    We can filter a dataframe by using a boolean mask. This mask is created by first selecting a column and
    applying some relational operators on it. This then gives us a boolean series which can then be used to filter
    out certain rows from a dataframe.

```python
ipl = pd.read_csv('ipl-matches.csv')
ipl.head(2)
```

#### `Find all the final winners.`

```python
mask = ipl['MatchNumber'] == 'Final'
winners = ipl[mask]['WinningTeam'].tolist()
winners
```

#### `How many super over finishes have occured?`

```python
mask = ipl['SuperOver'] == 'Y'
mask.sum()
```

#### `How many matches has CSK won in kolkata?`

```python
ipl
```

```python
mask1 = ipl['WinningTeam'] == 'Chennai Super Kings'
mask2 = ipl['City'] == 'Kolkata'
mask = mask1 & mask2

mask.sum()
```

#### `Find how many times toss winner is match winner. Also find the percentage.`

```python
mask = ipl['TossWinner'] == ipl['WinningTeam']
ans = mask.sum()
```

```python
ans/(ipl.shape[0]) * 100
```

### `Write a function which returns the track records of two teams against each other`

    i.e The number of matches won by each team.

```python
def track_record(team_one, team_two):
    teams = ipl['Team1'].tolist()
    teams.extend(ipl['Team2'].tolist())
    teams = set(teams) # list of all the unique teams
    
    if team_one not in teams or (team_two not in teams):
        return None
    else:
        wins_one = (ipl['WinningTeam'] == team_one).sum() # matches won by team one
        wins_two = (ipl['WinningTeam'] == team_two).sum() # matches won by team two
        
    return pd.Series({team_one : wins_one, team_two : wins_two})
```

```python
track_record('Chennai Super Kings','Deccan Chargers').plot(kind = 'bar')
```
