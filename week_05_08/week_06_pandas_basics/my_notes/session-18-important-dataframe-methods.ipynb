 ## `sort_values()`

    The sort_values() method is used to sort a series or a dataframe based on a particular column or a group
    of columns.
    
    We can also specify whether we want to sort by ascending or by descending order.

```python
s = pd.Series([1, 23, 45, 2, 3, 4, 5, -90, np.nan, np.nan])
s
```

```python
s.sort_values() # By default it sorts in the ascending order
```

`We can also sort by descending order by setting ascending = False.`

```python
s.sort_values(ascending=False)
```

`We can also do the sorting inplace. This would modify the original sequence permanently.`

```python
s.sort_values(ascending=False, inplace=True)
s
```

`We can also decide the position of missing values. This is done be setting the na_position parameter.`

```python
s.sort_values(ascending=False, na_position='first')
```

`We can also use sort_values() in the dataframes. We may specify a single column or a group of columns.`

```python
chicago = pd.read_csv('https://raw.githubusercontent.com/sameerjha462000/datasets/main/chicago.csv')
chicago = chicago.sample(chicago.index.size) # randomly shuffled the dataframe as it was sorted on Name column.
chicago
```

```python
chicago.sort_values(by='Name')
```

```python
chicago.sort_values(by = ['Name', 'Position Title']) # first sorts on Name and then on Position Title both in asc order.
```

```python
# chicago.sort_values(by = ['Position Title', 'Name'], ascending = [False, False]) # descending on both -- Name and Position Title
# chicago.sort_values(by = ['Position Title', 'Name'], ascending = [True, False]) # ascending on Position Title and descending on Name
chicago.sort_values(by = ['Position Title', 'Name'], ascending = [False, True]) # ascending on Name and descending on Position Title
# chicago.sort_values(by = ['Position Title', 'Name'], ascending = [True, True]) # default -- both ascending order
```

## rank()

    This method is used to give order to the series and dataframe values. By default, the smallest value
    is given the first position. If we want to change the default behaviour then we need to set 
    ascending = False.
    
    The values which are identical are given the same rank.

```python
names = pd.Series(['Abhishek', 'Amrusha', 'Priyanka', 'Sameer', 'Ibrahim', 'Daniyaal'])
names
```

```python
names.rank()
```

```python
names_dataframes = pd.DataFrame({'name' : ['Abhishek', 'Amrusha', 'Priyanka', 'Sameer', 'Ibrahim', 'Daniyaal', 'Abhishek']})
names_dataframes
```

`The identical values are given the same rank as seen below.`

```python
names_dataframes['rank'] = names_dataframes['name'].rank()
names_dataframes
```

## `set_index()` & `reset_index()`

    The set_index() is used to use an existing column as the label for the dataframe.
    The reset_index() is used to use the pandas default indexing.
    
    The set_index() drops the current index of the dataframe.
    The reset_index() does NOT drop the current index of the dataframe. Instead it converts the current index
    into a column of the dataframe.

```python
pokemon = pd.read_csv('https://raw.githubusercontent.com/sameerjha462000/datasets/main/pokemon.csv')
pokemon
```

```python
pokemon = pokemon.set_index('Name') # It drops the current index of the dataframe.
pokemon
```

```python
pokemon = pokemon.reset_index() # It does not drop the current index of the dataframe Instead it makes it into a new column
pokemon
```

    These functions have also a drop parameter. The reset_index() has drop parameter set to False by default due
    to which it does not drop the current index. Similarily, the set_index() has drop parameter set to True by
    default due to which it drops the column to be used as index. We can change their default behaviour by providing the value
    of drop parameter.

### `series to dataframe using reset_index() and set_index() function`

```python
pokemon = pd.read_csv('https://raw.githubusercontent.com/sameerjha462000/datasets/main/pokemon.csv', index_col = 'Name').squeeze(True)
pokemon
```

```python
pokemon_dataframe = pokemon.reset_index().set_index('Name')
pokemon_dataframe
```

# `rename`

    The rename() method is used to rename the index and columns of a dataframe.
    
    We either do this providing index or columns arguments. Or we may use the mapper argument and axis to
    specify what do we want to change.
    
    The axis = 0 refers to index and axis = 1 refers to columns.

```python
movies.head(2)
```

```python
movies.rename(mapper = lambda x : x.upper(), axis = 1).head(2)
```

```python
movies.rename(columns = lambda x : x.upper()).head(2)
```

```python
movies.set_index('title_x').rename(mapper = lambda x : x.upper(), axis = 0)
```

```python
movies.set_index('title_x').rename(index = lambda x:x.upper())
```

```python
s.nunique(dropna = False)
```

#### `fiilna() and dropna() in dataframes `

```python
employees = pd.read_csv('https://raw.githubusercontent.com/sameerjha462000/datasets/main/employees.csv')
employees
```

`By default when we use dropna() on dataframes it drops every row which has even a single missing value in any column.`

```python
employees.dropna()
```

`To change the default behaviour we can use the how parameter. By setting how = 'all' we will only these rows which has all missing values in all the columns.`

```python
employees.dropna(how = 'all')
```

`Since the above dataframe did not have a single row in which all the column values were missing, none of the rows were dropped.`

```python
students = pd.DataFrame(
    {
        'name':['nitish','ankit','rupesh',np.nan,'mrityunjay',np.nan,'rishabh',np.nan,'aditya',np.nan],
        'college':['bit','iit','vit',np.nan,np.nan,'vlsi','ssit',np.nan,np.nan,'git'],
        'branch':['eee','it','cse',np.nan,'me','ce','civ','cse','bio',np.nan],
        'cgpa':[6.66,8.25,6.41,np.nan,5.6,9.0,7.4,10,7.4,np.nan],
        'package':[4,5,6,np.nan,6,7,8,9,np.nan,np.nan]

    }
)

students
```

```python
students.dropna(how = 'all')
```

```python
students.dropna(how = 'any') # This is the default value of how parameter
```

`We can also drop rows on the basis of certain columns i.e drop if we found a missing value in that column of dataframe.`

```python
students
```

```python
students.dropna(subset=['name']) # drop all the rows where name is missing -- 3, 5, 7, 9
```

```python
students.dropna(subset = ['name', 'college']) # drop all the rows where name OR college is missing
```

`If we want to remove all the rows where name AND college are missing, then we can specify how = 'all'.`

```python
students
```

```python
students.dropna(subset=['name', 'college'], how='all') # drops all the rows where name AND college are missing -- 3, 7
```
