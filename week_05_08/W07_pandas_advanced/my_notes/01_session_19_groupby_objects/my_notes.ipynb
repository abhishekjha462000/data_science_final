# `groupby object`

```python
import numpy as np
import pandas as pd
```

```python
imdb = pd.read_csv('imdb-top-1000.csv')
imdb
```

`The groupby object is used to group the rows of a column on the basis of a values of a column. This column must be a categorical column. For example in the above dataframe, we have "Genre" column which is a categorical column. Hence, we may group by Genre(jawn-ruhh).`

```python
genres = imdb.groupby('Genre')
genres
```

```python
genres.__class__.__name__ # This is a DataFrameGroupBy object
```

`We can see the number of groups inside this groupby object by using the python's len function.`

```python
len(genres)
```

## `size()`

`We can get the number of rows inside each individual group by using the size() method.`

```python
genres.size()
```

## `head()`
    
    This method returns the 5 rows of each individual group. 
    If a group has less than 5 rows then all the rows would be shown.

```python
genres.head()
```

```python
genres.tail() # shows 5 rows from the bottom of each genre. If a genre has less than 5 rows, then all the rows would be shown
```

```python
genres.sample() # selects 1 random row from each group
```

## `first()`

    The first() method is used to give the first row of each group in our groupby object.

```python
genres.first()
```

## `last()`

    The last() method is used to give last row of each group from our groupby object.

```python
genres.last()
```

## `nth()`

    The nth() method takes in a number n.
    It then returns us the nth row from each group of our groupby object.
    If a particular row does not have nth entry, then we do not see any row from that particular group.

```python
genres.nth(3) # We do not have any family, fantasy, thriller Genre in this result as they didn't have 3rd row in them.
```

## `replace parameter in groupby sample method`

    When we try to use sample method with a value n which is > 1, then we need to mention replace = True as it might be a
    case that certain groups have less than n entries and thus we would not be able to pull n rows from them without 
    replacement.

```python
genres.sample(2, replace = True) # Thriller genre had only 1 row. So it got repeated twice.
```

## `agg()`

    The agg() method is used to apply aggregate function on columns of groupby object.

```python
genres.agg(
    {
        'Runtime':'mean',
        'IMDB_Rating':'mean',
        'No_of_Votes':'sum',
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

```python
genres.agg(['min','max','sum'])
```

```python
genres.agg(
    {
        'Runtime':['min','mean'],
        'IMDB_Rating':'mean',
        'No_of_Votes':['sum','max'],
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

## `looping over groupby objects`

    When we loop over group by object, then we get a tuple at each step of iteration. The first component of the
    tuple is the name of the group and the second component of the tuple is the group itself.

```python
genres
```

```python
for group_name, group in genres:
    print(group_name)
```

```python
for group_name, group in genres:
    print(group.head(1))
    print('\n' * 5)
```

## `apply()`

    The apply() method is used to apply a function on each group of a groupby object.

```python
genres
```

```python
genres.apply(lambda x : x)
```

```python
genres.apply(lambda x : x[['Released_Year', 'Genre']]) # x is each individual group
```

### `find the gross revenue in each genre`

```python
genres.apply(lambda x : x['Gross'].sum())
```

### `find the average runtime of each genre`

```python
genres.apply(lambda x : x['Runtime'].mean())
```

### `find number of movies starting with A for each group`

```python
genres.apply(lambda x : x['Series_Title'].str.startswith('A').sum())
```

### `find ranking of each movie in the group according to IMDB score`


```python
def rank_movie(group):
    group['rank'] = group['IMDB_Rating'].rank(ascending = False)
    return group
```

```python
genres.apply(rank_movie)
```

## `ipl wala dataset`

```python
ipl = pd.read_csv('deliveries.csv')
ipl.head()
```

### `find the top 10 batsman in terms of runs`

```python
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending = False).head(10)
```

### `find the batsman with max no of sixes`

```python
mask = ipl['batsman_runs'] == 6
sixes = ipl[mask]
```

```python
sixes.groupby('batsman').size().sort_values(ascending = False).head(1).index[0]
```

### `find batsman with most number of 4's and 6's in last 5 overs`

```python
# first find all the rows which correspond to the last 5 overs.
mask1 = ipl["over"] > 15

# find all the rows with 4's and 6's
mask2 = ipl['batsman_runs'].isin([4, 6])

mask = mask1 & mask2

df = ipl[mask]
```

```python
df.groupby('batsman').size().sort_values(ascending = False).head(1).index[0]
```

### `find V Kohli's record against all teams`

```python
v_kohli = ipl[ipl['batsman'] == 'V Kohli']
v_kohli.groupby('bowling_team')['batsman_runs'].sum().reset_index()
```

### `Create a function that can return the highest score of any batsman`

```python
def foo(batsman):
    mask = ipl['batsman'] == batsman
    
    df = ipl[mask]
    
    return df.groupby('match_id')['batsman_runs']\
.sum().sort_values(ascending = False).head(1).values[0]
```

```python
foo('V Kohli')
```
