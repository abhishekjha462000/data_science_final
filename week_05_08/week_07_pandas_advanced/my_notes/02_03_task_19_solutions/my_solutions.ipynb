### `Problem 1`
`Using groupby make groups using the Pclass column and find out the average age and total number of missing values in the Age column for every group.`


````python
import numpy as np
import pandas as pd
````

````python
titanic_df = pd.read_csv("train - train.csv")
titanic_df.head()
````

````python
classes = titanic_df.groupby('Pclass')
````

````python
# average age of each Pclass
classes['Age'].mean()
````

````python
# total number of missing values in the Age column for every group
classes.size() - classes['Age'].count()
````

### `Problem 2` 
`Using groupby make groups using the Pclass column and fill every group's Embarked column's missing values with the mode value of that group. After that, print every group's Embarked column's value counts in ascending order.`

````python
classes = titanic_df.groupby('Pclass')
````

````python
for group_name, group in classes:
    print(group_name, group['Embarked'].mode())
````

````python
def fill(group):
    group['Embarked'].fillna(group['Embarked'].mode(), inplace = True)
    return group
````

````python
df = classes.apply(fill)
df
````

````python
df['Embarked'].isnull().sum()
````

### `Problem 3:` 
`Make groups based on Embarked column. And for each of this embarked group, make another group based on Pclass and find out the average fare (round off up to 2 decimal places) for each Pclass for each group of Embarked.`

**Sample Output:**

```bash
{'C': {1: 105, 2: 25, 3: 11},
 'Q': {1: 90, 2: 12, 3: 11},
 'S': {1: 70, 2: 20, 3: 15}}
```

````python
embarked_groups = titanic_df.groupby('Embarked')
````

````python
for group_name, group in embarked_groups:
    print(group_name)
````

````python
for group_name, group in embarked_groups:
    # for each of the group, make groups based on Pclass
    pclass_groups = group.groupby('Pclass')
    average_fares = pclass_groups['Fare'].mean() # This would return us a series
    
    print(dict(average_fares))
    print('\n')
````

````python
# final code
ans = {}
for group_name, group in embarked_groups:
    # for each of the group, make groups based on Pclass
    pclass_groups = group.groupby('Pclass')
    average_fares = pclass_groups['Fare'].mean() # This would return us a series
    
    ans[group_name] = dict(average_fares)
    
ans
````
