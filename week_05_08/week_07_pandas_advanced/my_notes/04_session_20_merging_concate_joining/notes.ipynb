# `concat`

    We can concat the dataframes on top of each other using the top-level pd.concat() function

```python
import numpy as np
import pandas as pd
```

```python
nov = pd.read_csv('reg-month1.csv')
dec = pd.read_csv('reg-month2.csv')
nov
```

```python
dec
```

`We can merge nov and dec on top of each other.`

```python
pd.concat([nov, dec])
```

#### `Doing this however does not drop the indices of the dec dataframe. In order to drop it, we use ignore_index parameter.`

```python
pd.concat([nov, dec], ignore_index=True)
```

# `multi index dataframe`

We can also provide keys while we are concatenating two or more dataframes. This would however result in a
multiindex dataframe. A multi index dataframe is a dataframe whose index consists of more than 1 columns.

```python
multi_df = pd.concat([nov, dec], keys=['Nov', 'Dec'])
multi_df
```

```python
multi_df.loc['Nov']
```

```python
multi_df.loc['Dec']
```

```python
multi_df.loc[('Nov', 1)]
```

# `joins`

    To join 2 or more dataframes on the basis of a common column, we use the merge() method.
    
    The dataframe on the left hand side of the . is considered as the left table.
    The dataframe on the right hand side of the . is considered as the right table.


```python
courses = pd.read_csv('courses.csv')
regs = pd.concat([nov, dec], ignore_index=True)
students = pd.read_csv('students.csv')
```

```python
courses.head()
```

```python
students.head()
```

```python
regs.head()
```

### `Inner join`

    In case of inner join, we first find all the common values of the common values in both the dataframes.
    Then we show all the corresponding rows.

```python
np.intersect1d(students['student_id'].values, regs['student_id'].values)
```

`This means that when we will perform the inner join on the students and regs tables on the basis of student_id column then we will see the results where the value of student_id lies in the above array.`

```python
merged_df = students.merge(regs, how = 'inner', on = 'student_id')
merged_df.head(5)
```

```python
np.unique(merged_df['student_id'].values)
```

## `left join`

    When we do left join of two tables on the basis of a column, then all the values of common column are taken into
    consideration while creating the resulting table. It does not matter if the value is in the right table or not

```python
courses
```

```python
regs.head(10)
```

```python
courses['course_id'].values
```

```python
np.unique(regs['course_id'].values)
```

`From the above result it is clear that the course id 11 and 12 are not present in the regs. When we will do left join of
courses with regs, then course id 11 and 12 would also be included in the resulting dataframe and the student_id would be filled with missing value as no student has registered in these courses.`

```python
courses.merge(regs, how = 'left', on = 'course_id')
```

## `Problem 5`

    Find students who enrolled in both the months.

```python
student_ids_nov = nov['student_id'].values
student_ids_nov
```

```python
student_ids_dec = dec['student_id'].values
student_ids_dec
```

```python
ids_loyal_students = np.intersect1d(student_ids_nov, student_ids_dec)
ids_loyal_students
```

`Now we need to find the name of these students.`

```python
names = students[students['student_id'].isin(ids_loyal_students)]['name'].values.tolist()
names
```

## `find top 3 students who did most enrollments.`

```python
students_list = regs.merge(students, on = 'student_id').groupby(['student_id', 'name']).size().sort_values(ascending = False)\
.head(3).index.tolist()
students_list
```

```python
names_of_students_enrollments = [name for id, name in students_list]
names_of_students_enrollments
```

## `find top 3 students who spent most amount of money on courses`

```python
money_students = regs.merge(students, on = 'student_id').merge(courses, on = 'course_id')\
.groupby(['student_id', 'name'])['price'].sum().sort_values(ascending = False).head(3).index.tolist()
money_students
```

```python
ans = [name for id, name in money_students]
```

```python
ans
```

# `IPL wala dataset`

```python
matches = pd.read_csv('matches.csv')
delivery = pd.read_csv('deliveries.csv')
matches.rename(columns = {
    'id' : 'match_id'
}, inplace = True)
```

```python
matches.head()
```

```python
delivery.head(3)
```

## `find top 3 stadiums with highest sixes/matches ratio.`

First let us find the number of matches conducted in each stadium

```python
stadium_vs_no_of_matches = matches.groupby('venue').size()
stadium_vs_no_of_matches
```

#### `Now let us find the number of sixes in each stadium`

```python
mask = delivery['batsman_runs'] == 6
stadium_vs_no_of_sixes = matches.merge(delivery[mask], on = 'match_id').groupby('venue').size()
stadium_vs_no_of_sixes
```

```python
ans = stadium_vs_no_of_sixes/stadium_vs_no_of_matches
```

```python
ans
```

```python
ans.sort_values(ascending = False).head(3).index.tolist()
```
