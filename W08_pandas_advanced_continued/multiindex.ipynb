# Multiindex

A multiindex is an index object which has more than 1 dimensions.

```python
subjects = ['CSE', 'MCE', 'ECE']
years = [2020, 2021, 2022]

pd.MultiIndex.from_product([subjects, years])
```

```python
tuples = [('CSE', 2020),
            ('CSE', 2021),
            ('CSE', 2022),
            ('MCE', 2020),
            ('MCE', 2021),
            ('MCE', 2022),
            ('ECE', 2020),
            ('ECE', 2021),
            ('ECE', 2022)]
```

```python
mul_index = pd.MultiIndex.from_tuples(tuples)
```

```python
mul_index
```

```python
dir(mul_index)
```

```python
mul_index.levels[1]
```

```python
mul_index.swaplevel()
```

```python
mul
```

```python
mul.swaplevel().loc[('Argentina')]
```

# stack() and unstack()

```python
mul
```

```python
index_val = [('cse',2019),('cse',2020),('cse',2021),('cse',2022),('ece',2019),('ece',2020),('ece',2021),('ece',2022)]
multi_index = pd.MultiIndex.from_tuples(index_val)
multi_index
```

```python
s = pd.Series([1,2,3,4,5,6,7,8],index=multi_index)
s
```

```python
s.unstack()
```

```python
s.unstack().unstack()
```

```python
i = pd.MultiIndex.from_product([subjects, years, ['male', 'female']])
i
```

```python
s = pd.Series(range(1, 19), index = i)
s
```

```python
s.unstack()
```

```python
ss = s.unstack().unstack()
ss
```

```python
ss.columns.__class__.__name__
```

```python
sss = ss.unstack()
sss
```

```python
sss.loc[('female', 2022, 'ECE')]
```

```python
ss
```

```python
ss.stack()
```

```python
ss.stack().stack()
```

```python
list(ss.stack().stack().loc[[('ECE', 2020, 'female'), ('ECE', 2021, 'male')]].values)
```

```python
list(ss.stack().stack().swaplevel(0, 1).loc[2020].swaplevel().loc['female'].values)
```
