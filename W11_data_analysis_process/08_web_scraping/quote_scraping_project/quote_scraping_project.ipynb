```python
texts = []
authors = []
tags = []
```

```python
for quote in quotes:
    text = quote.find('span', class_ = 'text').text.replace('\t', '').strip('\n')
    author = quote.find('small', class_ = 'author').text.replace('\t', '').strip('\n')
    quote_tags = [tag.text.replace('\t', '').strip('\n') for tag in quote.find_all('a', class_ = 'tag')]
    
    texts.append(text)
    authors.append(author)
    tags.append(', '.join(quote_tags))
```

# scrape all the pages of [https://quotes.toscrape.com/]

```python
quotes = []
authors = []
tags = []
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
```

```python
for page_no in range(1, 100000_0000):
    response = requests.get(f'https://quotes.toscrape.com/page/{page_no}/', headers = headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    cards = soup.find_all('div', class_ = 'quote')
    
    if len(cards) == 0:
        break
    
    for card in cards:
        quote = card.find('span', class_ = 'text').text.replace('\t', '').strip('\n')
        author = card.find('small', class_ = 'author').text.replace('\t', '').strip('\n')
        quote_tags = [tag.text.replace('\t', '').strip('\n') for tag in card.find_all('a', class_ = 'tag')]

        quotes.append(quote)
        authors.append(author)
        tags.append(', '.join(quote_tags))
```

```python
quotes_df = pd.DataFrame({'quote': quotes, 'author' : authors, 'tags' : tags})
quotes_df
```

```python
quotes_df.to_csv('quotes_to_scrape_scraped.csv', index = False)
```
