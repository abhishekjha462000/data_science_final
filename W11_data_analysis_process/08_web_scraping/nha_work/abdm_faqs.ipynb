```python
from selenium import webdriver
import time
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
```

```python
s = Service(path = 'https://github.com/sameerjha462000/data_science_final/blob/main/W11_data_analysis_process/08_web_scraping/chromedriver.exe')
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
```

```python
driver = webdriver.Chrome(service = s)

url = 'https://abdm.gov.in/faqs'
driver.get(url)

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
faqs = [button.text for button in soup.select('button.accordion-button.collapsed')]
# faqs


xpaths = ['//*[@id="label-tab-2"]', '//*[@id="label-tab-3"]', '//*[@id="label-tab-4"]', '//*[@id="label-tab-5"]', '//*[@id="label-tab-6"]',
          '//*[@id="label-tab-7"]', '//*[@id="label-tab-8"]', '//*[@id="label-tab-9"]']


for xpath in xpaths:
    driver.find_element(by = By.XPATH, value = xpath).click()
    time.sleep(1.5)

    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    new_faqs = [button.text for button in soup.select('button.accordion-button.collapsed')]

    faqs.extend(new_faqs)

faqs
```

```python
import pandas as pd
faqs_abdm = pd.DataFrame({'faqs' : faqs, 'updation_required' : 'No', 'deletion_required' : 'No'})
```
